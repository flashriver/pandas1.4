### 2. 练习
#### 【练习一】 现有一份关于美剧《权力的游戏》剧本的数据集，请解决以下问题：
#### （a）在所有的数据中，一共出现了多少人物？
#### （b）以单元格计数（即简单把一个单元格视作一句），谁说了最多的话？
#### （c）以单词计数，谁说了最多的单词？

```python
winter= pd.read_csv('data/Game_of_Thrones_Script.csv')
winter.head()
```

```python
winter.shape
```

```python
winter.Name.value_counts()
# (a)一共564个人物, (b)tyrion lannister说了1760句话
```

```python
# 使用正则表达式,将字母都替换掉, 然后剩下的就是分隔符,分隔符个数加一,就得到每句话的单词个数.
import re
winter['num_of_words']=winter.Sentence.apply(lambda x:len(re.sub('\w','',x)))
```

```python
winter.num_of_words.groupby(winter.Name).sum().sort_values(ascending=False)
# (c) tyrion lannister说了30259个单词 
```

#### 【练习二】现有一份关于科比的投篮数据集，请解决如下问题：
#### （a）哪种action_type和combined_shot_type的组合是最多的？
#### （b）在所有被记录的game_id中，遭遇到最多的opponent是一个支？

```python
kobe=pd.read_csv('data/Kobe_data.csv',index_col='shot_id')
kobe.head()
#index_col的作用是将某一列作为行索引
```

```python
kobe.shape
```

```python
(kobe.action_type+' & '+kobe.combined_shot_type).value_counts().head()
# (a)Jump Shot & Jump Shot最多,一共有18880次
```

```python
# 提取game_id,然后去重,然后与opponentjoin,得到game_id具有唯一性的一个df,然后统计opponent列的个数
oppenents=kobe[['game_id']].drop_duplicates().join(kobe[['opponent']],how='inner')
oppenents
```

```python
oppenents.opponent.value_counts().sort_values(ascending=False)
# (b)对马刺的次数最多, 有91次.
```

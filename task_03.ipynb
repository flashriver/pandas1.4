```python
# 多行索引时, 需传入一个list,而不是多个索引
df.loc[[1102,2304]]
```

```python
# 定义一个为两个list做笛卡尔乘积的函数
def list_product(list1,list2):
    lst=[]
    for a in list1:
        for b in list2:
            lst.append((a,b))
    return lst
```
